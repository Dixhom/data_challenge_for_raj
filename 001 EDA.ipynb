```python
# calculate the time spent by a user
# count of the seconds spent for websearch for each query
# then sum up the second counts
def time_spent(x):
    tmp = x.groupby('query_').querytime.apply(lambda y: (y.max() - y.min()).seconds)
    return sum(tmp) 

time_spent_sec = df.groupby('anonid').apply(time_spent)       
```

```python
time_spent_sec
```

```python
time_spent_sec.hist(bins=100)
```

```python
time_spent_sec.mean()
```
